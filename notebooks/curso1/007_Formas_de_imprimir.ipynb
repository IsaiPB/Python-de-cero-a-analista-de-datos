
# 1. Uso simple de print

```python
print( "Hola"+ "Mundo")
```

```python
palabra = "Feliz" # Variable que irá en print

print( "Hola" + "Mundo" + palabra) # Concatenar con operaciones (+)
```

```python
numero = 10
print( "Hola" + "Mundo" + numero) # No es posible concatenar string con enteros
```

```python
numero = 10
print( "Hola" + "Mundo" + str(numero)) # Concatenar string con enteros (usando str)
```

```python
print("Hola Mundo", numero, numero) # Concatenar string con enteros (usando ,)
```

# 2. Operador %

```python
nombre = "Diego"
print("Hola, %s!" % nombre)
```

```python
lista = [1,2,3,4,5]
print("Hola, %s!" % lista)
```

# 3. Format

```python
# Distintas formas de usar Format
mi_nombre = "Diego"
fecha = "10/10/2020"
lugar = "la luna"
print("Hola, {}! Hoy {} en {}.".format(mi_nombre, fecha, lugar))
print("Hola, {0}! Hoy {1} en {2}.".format(mi_nombre, fecha, lugar))
print("Hola, {nombre}! Hoy {fecha} en {lugar}.".format(nombre=mi_nombre, fecha=fecha, lugar=lugar))
```

# 4. f-strings

```python
nombre = "Diego"
fecha = "10/10/2020"
lugar = "la luna"
print(f"Hola, {nombre}! Hoy {fecha} en {lugar}.")
```

## 4.1 Ej. de Variables y f-strings en python: uso de decimales

```python
# Resolución de problema usando variables
# $F = 9/5C + 32$
c = 106
f = 9/5*c + 32
print(f"{c}°C son {f: .3f}°F")
```
