# 1. Convertir una cadena en lista

```python
frase = "En un lugar de la Mancha, de cuyo nombre no quiero acordarme"
type(frase)
```

```python
# Cada elemento de la cadena se vuelve un elemento de la lista

list(frase) # Convertir frase en lista
```

```python
# Poner toda la frase en una lista

list([frase])
```

# 2. Dividir una cadena en varias listas

```python
frase.split("e") # Dividir frase en palabras
```

```python
lista_1 = frase.split() # Dividir frase en palabras
lista_1
```

# 3. Unir elementos de una lista usando un conector

```python
lista_2 = "+".join(lista_1) # Unir elementos de la lista usando un conector
```
