# 1. if

```python
# Ejemplo de condición if

edad = 18
if edad >= 18:
    print("Es mayor de edad")
else:
    print("Es menor de edad")
```

# 2. else

```python
# Ejemplo de condición else

edad = 15
if edad >= 18:
    print("Es mayor de edad")
else:
    print("Es menor de edad")
```

# 3. elif

```python
# Ejemplo de elif

edad = 18
if edad > 18:
    print("Es mayor de edad")
elif edad == 18:
    print("Tienes 18 años")
else:
    print("Es menor de edad")
```

# 4. for

```python
# Iterar sobre una lista

for i in [1, 2, 3, 4, 5]:
    print(i)
```

```python
# Uso de función range

for i in range(5): #Excluyete (n-1)
    print(i)
```

```python
# Uso de función range de inicio a final

for i in range(2, 5): #Excluyete (n-1)
    print(i)
```

```python
# Saltos con la función range

for i in range(0, 10, 2): #Excluyente (n-1)
    print(i)
```

# 5. while

```python
i = 0
while i < 5:
    print(i)
    i += 1 #Incremento de i = i + 1
```

```python
# Error al no actualizar la variable de control

i = 0
while i < 5:
    print(i) # bucle infinito
```

# 6. List comprehension

```python
# Forma común de iterar en listas con for

colores = ["rojo", "verde", "azul"]
for color in colores:
    print(color)
```

```python
# Usar append para crear una nueva lista e iterar

nueva_lista = []
for color in colores:
    nueva_lista.append(color)
print(nueva_lista)
```

```python
# Obtener elementos que solo tinene la letra a

nueva_lista = []
for color in colores:
    if "a" in color:
        nueva_lista.append(color)
print(nueva_lista)
```

```python
# Usar List comprehension

nueva_lista = [color for color in colores if "a" in color]
print(nueva_lista)
```

```python
# convertir str a int

numeros = ["1", "2", "3", "4", "5"]
numeros = [int(numero) for numero in numeros]
print(numeros)
```
